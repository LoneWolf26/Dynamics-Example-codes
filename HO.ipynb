```python
# Import liblaries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Image format
plt.style.use('bmh')
figsize = (12, 8)
dpi = 300 
```

```python
# Constants and parameters
pi  = np.pi
A   = 1
g   = [2,1,0.1,0.01,0.0]
w0  = [1/4,1/8,1/20]
phi = [0, 1/2, 2, 4]
```

```python
# Time and number of time steps
T = 200
n = 1000
dt = T/float(n)
t = np.arange(-T, T, dt)
```

```python
# Plot for variation of angular velocity
plt.figure(figsize=(12,8), dpi=dpi)
plt.title(r"Changes in initial angular velocity")
for i in range(len(w0)):
    x_ud = A*np.cos(w0[i]*t + phi[0])
    x_d  = A*np.cos(w0[i]*t + phi[0])*np.exp(-g[3]*t)
    v_ud = -w0[i]*A*np.sin(w0[i]*t + phi[0])
    v_d  = A*(-g[3])*x_d + v_ud*np.exp(-g[3]*t)
    
    # Normalize
    x_d  = x_d/np.max(abs(x_d))
    v_d  = v_d/np.max(abs(v_d))
    x_ud = x_ud/np.max(abs(x_ud))
    v_ud = v_ud/np.max(abs(v_ud))

    color = cm.YlGnBu(float(i + 1)/len(w0)) # Creating colormap
    
    plt.subplot(221)
    plt.title(r"Undamped")
    plt.plot(t,x_ud,c=color)
    plt.ylabel(r"$x(t)/L$")
    plt.xlabel(r"$t$, [s]")
    plt.grid(True)
    
    plt.subplot(222)
    plt.title(r"Damped")
    plt.plot(t,x_d,c=color,label=r"$\omega_0=%.3fs^{-1}$"%(w0[i]))
    plt.ylabel(r"$x(t)/L$")
    plt.xlabel(r"$t$, [s]")
    plt.legend()
    plt.grid(True)
    
    plt.subplot(223)
    plt.title(r"Undamped Phase-Diagram")
    plt.plot(x_ud,v_ud,c=color)
    plt.xlabel(r"$x(t)/L$")
    plt.ylabel(r"$v(t)/(L/s)$")
    plt.grid(True)

    plt.subplot(224)
    plt.title(r"Damped Phase-Diagram")
    plt.plot(x_d,v_d,c=color)
    plt.xlabel(r"$x(t)/L$")
    plt.ylabel(r"$v(t)/(L/s)$")
    plt.grid(True)
    
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Plot for variation of dampling ratio
plt.figure(figsize=(12,4), dpi=dpi)
plt.title(r"Changes in dampning ratio")
for j in range(len(g)):
    x_ud = A*np.cos(w0[0]*t + phi[0])
    x_d  = A*np.cos(w0[0]*t + phi[0])*np.exp(-g[j]*t)
    v_ud = -w0[0]*A*np.sin(w0[0]*t + phi[0])
    v_d  = A*(-g[j])*x_d + v_ud*np.exp(-g[j]*t)
    
    # Normalize
    x_d = x_d/np.max(abs(x_d))
    v_d = v_d/np.max(abs(v_d))

    color = cm.YlGnBu(float(j + 1)/len(g)) # Creating colormap
    
    plt.subplot(121)
    plt.title(r"Different dampning ratios")
    plt.plot(t,x_d,c=color,label=r"$\gamma=%.2f$"%(g[j]))
    plt.ylabel(r"$x(t)/L$")
    plt.xlabel(r"$t$, [s]")
    plt.legend()
    plt.xlim([-200, -150])
    plt.grid(True)

    plt.subplot(122)
    plt.title(r"Phase-Diagram")
    plt.plot(x_d,v_d,c=color)
    plt.xlabel(r"$x(t)/L$")
    plt.ylabel(r"$v(t)/(L/s)$")
    plt.grid(True)

plt.tight_layout()
plt.show()
plt.close()
```

```python
plt.figure(figsize=(12,8), dpi=dpi)
plt.title(r"Changes in phase")
for k in range(len(phi)):
    x_ud = A*np.cos(w0[2]*t + phi[k])
    x_d  = A*np.cos(w0[2]*t + phi[k])*np.exp(-g[3]*t)
    v_ud = -w0[2]*A*np.sin(w0[2]*t + phi[k])
    v_d  = A*(-g[3])*x_d + v_ud*np.exp(-g[3]*t)
    
    # Normalize
    x_d  = x_d/np.max(abs(x_d))
    v_d  = v_d/np.max(abs(v_d))
    x_ud = x_ud/np.max(abs(x_ud))
    v_ud = v_ud/np.max(abs(v_ud))

    color = cm.YlGnBu(float(k + 1)/len(phi)) # Creating colormap
    
    plt.subplot(221)
    plt.title(r"Undamped")
    plt.plot(t,x_ud,c=color)
    plt.ylabel(r"$x(t)/L$")
    plt.xlabel(r"$t$, [s]")
    plt.grid(True)
    
    plt.subplot(222)
    plt.title(r"Damped")
    plt.plot(t,x_d,c=color,label=r"$\phi=%.3f$"%(phi[k]))
    plt.ylabel(r"$x(t)/L$")
    plt.xlabel(r"$t$, [s]")
    plt.legend()
    plt.grid(True)
    
    plt.subplot(223)
    plt.title(r"Undamped Phase-Diagram")
    plt.plot(x_ud,v_ud,c=color)
    plt.xlabel(r"$x(t)/L$")
    plt.ylabel(r"$v(t)/(L/s)$")
    plt.grid(True)

    plt.subplot(224)
    plt.title(r"Damped Phase-Diagram")
    plt.plot(x_d,v_d,c=color)
    plt.xlabel(r"$x(t)/L$")
    plt.ylabel(r"$v(t)/(L/s)$")
    plt.grid(True)
    
plt.tight_layout()
plt.show()
plt.close()
```
