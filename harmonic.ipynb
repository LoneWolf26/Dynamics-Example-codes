```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
g = 0.01
w_o = 0.15
```

```python
t    = np.linspace(-200,200,400)
x_ud = A*np.cos(w_o*t)
x_d  = A*np.cos(w_o*t)*np.exp(-g*t)
v_ud = -w_o*A*np.sin(w_o*t)
v_d  = A*(-g)*x_d + v_ud*np.exp(-g*t)

plt.title(r"Undamped")
plt.plot(t,x_ud)
plt.show()
plt.close()

plt.title(r"Damped")
plt.plot(t,x_d)
plt.show()
plt.close()

plt.title(r"Undamped Phase-Diagram")
plt.plot(x_ud,v_ud)
plt.show()
plt.close()

plt.title(r"Damped Phase-Diagram")
plt.plot(x_d,v_d)
plt.show()
plt.close()
```
